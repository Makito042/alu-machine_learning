## CO2 Distribution Plot

```python
# YOUR CODE HERE
plt.hist(df["co2"], bins=20, color="#ADD8E6")
plt.xlabel("CO2")
plt.ylabel("Frequency")
plt.title("CO2 Distribution")
plt.tight_layout()
plt.savefig("./plots/co2_hist_plot.png")
plt.show()
```

## Interpretation

- Most vehicles emit CO2 levels in the range of 300-400 units, with a peak around 350 units.
- The distribution is slightly right-skewed, meaning fewer vehicles emit very high levels of CO2.
- CO2 emissions range from around 0 to 700 units, with very few vehicles at the extremes (below 100 or above 600 units).
- The majority of vehicles fall in the mid-range of CO2 emissions, with over 600 vehicles producing around 300-350 units.

# Fuel Efficiency Distribution Plot

```python
# YOUR CODE HERE
plt.hist(df["feScore"], bins=10, color="green")
plt.xlabel("FE Score")
plt.ylabel("Frequency")
plt.title("Fuel Efficiency Distribution")
plt.tight_layout()
plt.savefig("./plots/fescore_hist_plot.png")
plt.show()
```

## Interpretation

- **Distribution:** The distribution is approximately normal (bell-shaped), centered around an FE score of 5 to 6, meaning most vehicles have a mid-range fuel efficiency.
- **Range:** FE scores range from 0 to 10, with the majority of vehicles scoring between 3 and 8, indicating a spread across different levels of fuel efficiency.
- **Peak:** The highest frequency of vehicles falls in the 5 to 6 FE score range, with over 1000 vehicles in this category, suggesting that this is the most common fuel efficiency level.
- **Symmetry:** The histogram is fairly symmetrical, indicating that high and low FE scores are equally uncommon, with most vehicles clustering around the middle range.

```python
# YOUR CODE HERE
plt.hist(df["ghgScore"], bins=10, color="green")
plt.xlabel("GHG Score")
plt.ylabel("Frequency")
plt.title("Greenhouse-Gas Score Distribution")
plt.tight_layout()
plt.savefig("./plots/ghgscore_hist_plot.png")
plt.show()
```

## Interpretation

- **Peak Frequency:** The highest number of vehicles falls within a GHG score of 5, with over 1,000 vehicles, indicating that the majority of vehicles have moderate greenhouse gas emissions.

- **Left Skew:** There is a noticeable left tail in the distribution, with a small number of vehicles scoring between 1 and 3, implying that very few vehicles are highly inefficient.

- **Decline in High Scores:** After GHG score 7, there is a sharp decline in the number of vehicles with higher scores (above 8), suggesting that very few vehicles are exceptionally eco-friendly.

- **Symmetry:** The distribution is nearly symmetrical, with similar frequencies on either side of the peak score (around 6), though slightly more vehicles score below 6 than above.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix
```

```python
custom_cmap = LinearSegmentedColormap.from_list(
    "custom_cmap",
    ("#c55d42", "#FFFFFF", "#458396")
)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, cmap=custom_cmap, vmin=-1, vmax=1)
plt.xticks(rotation=45, ha="right")
plt.savefig("./plots/fuel_efficiency_heatmap.png")
plt.show()
```

## Interpretation

The heatmap above display the corelation of different columns in the dataframe. Correlation values range from -1 to 1, where:

- 1 means perfect positive correlation (as one variable increases, so does the other),
- -1 means perfect negative correlation (as one variable increases, the other decreases),
- 0 means no correlation.

### My Observations

- City, Highway, Combined (comb), and Fuel Efficiency (fscore) are highly correlated (all above 0.9). This means that vehicles with better city fuel efficiency also tend to have better highway and combined fuel efficiency, as well as higher overall fuel scores.

- Both CO2 emissions and ghgScore are negatively correlated with city, highway, and combined fuel efficiency (correlations around -0.9). This indicates that vehicles with higher fuel efficiency tend to produce lower CO2 emissions and have better greenhouse gas scores.

- Cylinders, engine displacement (disp1), show a moderate negative correlation with fuel efficiency measures (around -0.7 to -0.8). Larger engines and more cylinders tend to result in lower fuel efficiency and higher emissions.
